## Creation of a DataAplose object

```python
from pathlib import Path

import matplotlib.pyplot as plt
from pandas import Timedelta

from disclose.dataclass.data_aplose import DataAplose

config = {
    "detection_file": Path(r"../../user_case/resource/APOCADO_results.csv"),
    "timebin_new": Timedelta("60s"),
    "filename_format": "%Y-%m-%dT%H-%M-%S",
}

data = DataAplose.from_dict(config)
```

## Barplot - All annotators / Same label

```python
fig, ax = plt.subplots(1, 1)
bin_size = Timedelta("1h")
ticks = Timedelta("2h")
fmt = "%H:%M"
ax = data.set_ax(ax=ax, x_ticks_res=ticks, date_format=fmt)

data.plot(
    mode="histogram",
    annotator=data.annotators,
    label=data.labels[0],
    ax=ax,
    bin_size=bin_size,
    season=False,
)
plt.show()
```

## Barplot - One annotator / All labels

```python
fig, ax = plt.subplots(1, 1)
bin_size = Timedelta("1h")
ticks = Timedelta("2h")
fmt = "%H:%M"
ax = data.set_ax(ax=ax, x_ticks_res=ticks, date_format=fmt)

data.plot(
    mode="histogram",
    annotator=data.annotators[0],
    label=data.labels,
    ax=ax,
    bin_size=bin_size,
    season=False,
)
plt.show()
```
