## Creation of a DataAplose object

```python
from pathlib import Path
from pandas import Timedelta
import matplotlib.pyplot as plt

from post_processing.dataclass.data_aplose import DataAplose

yaml_file = Path(r"_static/APOCADO_yaml.yml")
data = DataAplose.from_yaml(file=yaml_file)
```

## Barplot - All annotators / Same label

```python
fig, ax = plt.subplots(1, 1)
bin_size = Timedelta("1h")
ticks = Timedelta("2h")
fmt = "%H:%M"
ax = data.set_ax(ax=ax, x_ticks_res=ticks, date_format=fmt)

data.plot(
    mode="histogram",
    annotator=data.annotators,
    label=data.labels[0],
    ax=ax,
    bin_size=bin_size,
    season=False,
)
plt.show()
```

## Barplot - One annotator / All labels

```python
fig, ax = plt.subplots(1, 1)
bin_size = Timedelta("1h")
ticks = Timedelta("2h")
fmt = "%H:%M"
ax = data.set_ax(ax=ax, x_ticks_res=ticks, date_format=fmt)

data.plot(
    mode="histogram",
    annotator=data.annotators[0],
    label=data.labels,
    ax=ax,
    bin_size=bin_size,
    season=False,
)
plt.show()
```
